```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
pd.set_option('display.max_columns', 100)
pd.set_option('display.max_rows', 100)
np.set_printoptions(suppress=True)
```

```python
df = pd.read_csv('Berlin.csv')
```

```python
print(df.shape)
df.head()
```

```python
# Get ratios of unique values to total rows,
# for high cardinality checking purposes (discarding nulls)

rats = dict()
for c in df.columns:
    n = df[c].nunique()
    t = df[c].notna().sum()
    p = n / t if t > 0 else 1
    rats[c] = p

rats = sorted(rats.items(), key=lambda x: x[1])
x, y = zip(*rats)
plt.plot(x, y)
plt.title('Ratios of Unique to Total Non-Null Values')
plt.show()

#print(sorted(rats.items(), key=lambda x: x[1]))
```

```python
# note any features that have >80% cardinality

# not pulling columns it should, not sure why

hicard_cols = [x[i] for i in range(len(x)) if y[i] > 0.8]
print(hicard_cols)
print(df['id'].nunique() / df['id'].notna().sum())
```

```python
# do the same for null values

nuls = dict()
for c in df.columns:
    n = df[c].isna().sum()
    t = df.shape[0]
    p = n / t
    nuls[c] = p

nuls = sorted(nuls.items(), key=lambda x: x[1])
x, y = zip(*nuls)
plt.plot(x, y)
plt.title('Ratio of NA Values per Feature')
plt.show()
```

```python
# note any features that have >80% nulls

hinull_cols = [x[i] for i in range(len(x)) if y[i] > 0.8]
print(hinull_cols)
```

```python
drop_cols = sorted(list(set(hinull_cols + hicard_cols)))
print(drop_cols)
```

```python
df2 = df.drop(drop_cols, axis=1)
```
